```python
import pandas as pd
```

```python
articles_links = pd.read_csv('../data/raw/articles_links.csv', encoding='utf-8')
raw_medical_dataset = pd.read_csv('../data/raw/raw_medical_dataset.csv', encoding='utf-8')
```

```python
print(f"Số lượng link duy nhất trong all_nhom_benh_articles: {len(articles_links['link'].unique())}")
print(f"Số lượng link duy nhất trong raw_dataset: {len(raw_medical_dataset['link'].unique())}")
```

```python
duplicate_links = articles_links[articles_links['link'].duplicated(keep=False)]
display(duplicate_links)
print(f"Số lượng link bị trùng lặp trong dfa: {len(duplicate_links)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

category_counts = articles_links['category'].value_counts()

plt.figure(figsize=(12, 7))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('Distribution of Categories in articles_links', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Articles', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
